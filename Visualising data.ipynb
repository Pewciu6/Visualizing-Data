```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/kaggle/input/life-expectancy-averaged-dataset/Life-Expectancy-Data-Averaged.csv')
df.head(3)
```

```python
df_Region = df.groupby(by='Region').mean(['Life_expectancy'])
df_Region.head(len(df_Region))
```

```python
plt.figure(figsize=(8,6))
plt.scatter(df_Region.index, df_Region["Life_expectancy"]) 
plt.title('Life Expectancy in a Region', fontsize=14)
plt.xlabel('Region', fontsize=12)
plt.ylabel('Life Expectancy', fontsize=12)
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.5)
plt.show() 
```

```python
df_Country = df.groupby(by='Country').mean('Life_expectancy')
df_Country.head(len(df_Country))
```

```python
plt.figure(figsize=(30,15))
plt.bar(df_Country.index, df_Country['Life_expectancy'])
plt.title('Life Expectancy in Country')
plt.xlabel('Country')
plt.ylabel('Life Expectancy')
plt.xticks(rotation=90)
plt.show()
```

```python
df_Country.sort_values(by='Life_expectancy', ascending=False).head(5)
```

```python
df_Country.sort_values(by='Life_expectancy', ascending=True).head(5)
```

```python
df_Alcohol = df.set_index('Alcohol_consumption')[['Life_expectancy']]
df_Alcohol.head(5)
```

```python
df_Alcohol.sort_index(inplace=True)
plt.figure(figsize=(10,6))
plt.plot(df_Alcohol.index, df_Alcohol['Life_expectancy'], marker='o', color='blue')
plt.title('Life Expectancy vs Alcohol_consumption')
plt.xlabel('Alcohol_consumption')
plt.ylabel('Life Expectancy')
plt.grid(True)
plt.show()
```
